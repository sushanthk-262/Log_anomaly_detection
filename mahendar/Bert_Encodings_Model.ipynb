```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
```

```python
# Load the Data
df = pd.read_csv('/content/data.csv', sep=',', quotechar='"', names=["text", "label"])
df = df[1:201]
df
```

```python
# Feature extraction
X = list(df['text'])
```

```python
# Load the BERT tokenizer and model
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
encodings = tokenizer(X[0:10], truncation=True, padding=True, return_tensors='pt')
with torch.no_grad():
    # Obtain embeddings from the BERT model
    embeddings = model(**encodings).last_hidden_state
```

```python
mean_embeddings = embeddings.mean(dim=1)

# Convert to a NumPy array
mean_embeddings_np = mean_embeddings.numpy()

# Print the resulting embeddings
print(mean_embeddings_np)
```
